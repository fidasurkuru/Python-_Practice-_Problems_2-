```python
#1.Write a python program to find the largest of 3 numbers

num1 = 10
num2 = 20
num3 = 30

if (num1 >= num2) and (num1 >= num3):
   largest = num1
elif (num2 >= num1) and (num2 >= num3):
   largest = num2
else:
   largest = num3

print("The largest number is", largest)
```

```python
#3.Write a python program to find the count of numbers divisible by 8 from 1 to 1000
lower, upper = 1, 1000
count = 0

for num in range(lower, upper + 1):
    if num%8 == 0:
        count += 1

print(count)

```

```python
# 4.#Write a Python program to construct the following pattern, using anested for loop.

n = 9
for i in range(1,n+1):
    if i <= 5:
        for j in range(0, i):
            print('*', end="")
    else:
        for k in range(0, n-i+1):
            print('*', end="")
    print('')
```

```python
# 10. Write a Python program to check whether a number is prime or not.
x = 3
if x > 1:
    for i in range(2, int(x/2)+1):
        if (x % i) == 0:
            print(x, "is not a prime")
        break
    else:
        print(x, "is a prime")
else:
    print(x, "is not a prime")
```
